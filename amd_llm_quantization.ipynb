```python
import requests
import zipfile
import os
import subprocess

url = "https://www.xilinx.com/bin/public/openDownload?filename=quark-0.6.0.zip"
zip_filename = "quark-0.6.0+dba9ca364"
response = requests.get(url)
response.raise_for_status()

with open(zip_filename, "wb") as f:
    f.write(response.content)
with zipfile.ZipFile(zip_filename, "r") as zip_ref:
    zip_ref.extractall(".")

wheel_path = os.path.join("quark-0.6.0+dba9ca364", "quark-0.6.0+dba9ca364-py3-none-any.whl")
if not os.path.isfile(wheel_path):
    print("Error: whl not found.")
else:
    subprocess.check_call(["pip", "install", wheel_path])
    print("Installation complete.")
```
